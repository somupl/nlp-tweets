```python
import nltk
import pandas as pd
import numpy as np
import spacy
from spacy import displacy
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from itertools import combinations
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
%matplotlib inline
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
data = pd.read_csv('twitter-airline-sentiment/Tweets.csv')
text = data.pop('text')
label = data.pop('airline_sentiment')
```

```python
docs = [nlp(t) for t in text]
vecs = [d.vector for d in docs]
```

```python
X_train, X_test, y_train, y_test = train_test_split(vecs, label, test_size=0.30)
```

```python
simlarity = cosine_similarity(vecs)
```

```python
knn = KNeighborsClassifier(n_neighbors=3, algorithm='brute', metric = 'cosine')
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```
